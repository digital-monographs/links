```python
import matplotlib.pyplot as plt
import numpy as np
```

Content.

A chart.

```python
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='A chart')
ax.grid()

plt.show()
```
